```python
import pandas as pd
```

```python
variation = 'run_incremental'
scale = 10
partition = 10
eval_partitions = 2
query = 'q2'
log_file = f'logs/{variation}/scale={scale}/partition={partition}/query={query}/eval_partitions={eval_partitions}/latest.log'
```

```python
with open(log_file,'r') as f:
    logs = f.readlines()
```

```python
log_info = []
for line in logs:
    if 'func:' in line:
        time_instant = float(line.split()[0])
        event,function = line.split()[1].split('func:')[1].split(':')
        params = line.split()[2:]
        log_info.append({
            'timestamp': time_instant,
            'event': event,
            'function': function,
            'params': ' '.join(params)
        })
```

```python
df = pd.DataFrame(log_info)
df['time_delta'] = df['timestamp'].diff(periods=1)
df['cumulative'] = df['time_delta'].cumsum(axis=0)
```

```python
pd.set_option('display.max_rows', None)
```

```python
print("Total Time (in sec):",df['cumulative'].max())
```
